```python
#7.2 Analyze the monthly sum of confirmed cases,  deaths and recovered cases to understand the progression of the pandemic
```

```python
combined_df
```

```python
combined_df['Year month']=combined_df['Date'].dt.to_period('M')
```

```python
combined_df
```

```python
combined_df['Day'] = combined_df['Date'].dt.day
```

```python
combined_df['Year month'].value_counts()
```

```python
combined_df.isnull().sum()
```

```python
combined_df_country=combined_df.copy()
```

```python
combined_df_country
```

```python
combined_df_country['daily_confirmed']=combined_df_country.groupby(['Country/Region'])['Confirmed'].diff().fillna(0).clip(lower=0)
combined_df_country['daily_deaths']=combined_df_country.groupby(['Country/Region'])['Deaths'].diff().fillna(0).clip(lower=0)
combined_df_country['daily_recovered']=combined_df_country.groupby(['Country/Region'])['Recovered'].diff().fillna(0).clip(lower=0)
```

```python
combined_df_country
```

```python
monthly_sum=combined_df_country.groupby(['Country/Region', 'Year month'])[['daily_confirmed', 'daily_deaths','daily_recovered']].sum().reset_index()
```

```python
monthly_sum
```

```python
list_countries=['US', 'Italy', 'Brazil']
```

```python
combined_df_countries=monthly_sum[monthly_sum['Country/Region'].isin(list_countries)]
```

```python
combined_df_countries
```
